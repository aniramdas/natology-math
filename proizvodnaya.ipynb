```python
from scipy.misc import derivative
from math import sin,cos
def f(x): #первая производная
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
derivative(f, 1.0, dx=1e-6)
```

```python
x = 1 #самостоятельным решением производная имеет вид: 5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x)
5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x)
```

```python
def f(x): #вторая производная
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
derivative(f, 1.0, dx=1e-6, n=2)
```

```python
x = 1 #самостоятельным решением производная имеет вид 20*x**3 - 16*sin(2*x) - 9*cos(3*x + 3)
20*x**3 - 16*sin(2*x) - 9*cos(3*x + 3)
```

```python
import numpy as np
from math import sin,cos
import matplotlib.pyplot as plt

def f(x):
    return sin(5*(2*x + 1))
x = np.linspace(-5, 5)
```

```python
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.plot(x, f1x, label='first derivative')
plt.plot(x, f2x, label='second derivative')
plt.legend()
plt.show()
```
