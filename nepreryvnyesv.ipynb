```python
import numpy as np
import pandas as pd
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
mu = (1.0,2.0)
sigma = (0.5,0.75,1.0)
```

```python
for m in mu:
    for s in sigma:
        norm_rv = sts.norm(loc = m, scale = s)
        norm_rv.rvs(size = 10) # генерируемые 10 значений 
        norm_rv.pdf(3)
        x = np.linspace(0, 4, 100)
        pdf = norm_rv.pdf(x)
        plt.plot(x, pdf, label=f'mu={m}, sigma={s}')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend()
```

Как мы видим по графику, mu влияет на сдвиг по горизонтальной оси (х), а sigma - на ширину вершины.

```python
lam = (0.1, 0.25, 0.5, 0.75, 0.9)

for l in lam:
    expon_rv = sts.expon(l) 
    expon_rv.rvs(10)
    x = np.linspace(0,5, 100)
    pdf = expon_rv.pdf(x)
    plt.plot(x, pdf, label=f'lambda={l}')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend()
```

Лямбда влияет на сдвиг по горизонтальной оси и на расположение кривой (а точнее ее приближенность к оси х).

```python
n = (0.1, 1, 10, 100, 1000)

for nn in n:
    t_rv = sts.t(nn) 
    t_rv.rvs(10)
    x = np.linspace(-5,5, 100)
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf, label=f'n={nn}')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend()
```

Чем больше n, тем сильнее t-распределение стремится к нормальному.

```python
p_red = 54/(54+4+28)
p_yellow = 4/(54+4+28)
p_green = 28/(54+4+28)

i_red = -(p_red*np.log2(p_red)) 
i_yellow = -(p_yellow*np.log2(p_yellow)) 
i_green = -(p_green*np.log2(p_green))
```

```python
print(f'вероятность того, что светофор горит красным = {round(p_red, 4)}')
print(f'сообщение о том, что светофор красный, несет {round(i_red, 4)} бит')

print(f'вероятность того, что светофор горит желтым = {round(p_yellow, 4)}')
print(f'сообщение о том, что светофор желтый, несет {round(i_yellow, 4)} бит')

print(f'вероятность того, что светофор горит зеленым = {round(p_green, 4)}')
print(f'сообщение о том, что светофор зеленый, несет {round(i_green, 4)} бит')
```
