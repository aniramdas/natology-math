Уровень 1

```python
def f(x,c0,c1):
    return (c0*x[0]**2) + (c1*x[1]**2)
```

```python
c0, c1 = (1,1)
xn = np.array([100, 200])
yn = f(xn,c0, c1)
learning_rate = 0.1
```

```python
Y = {yn: xn}
Y
```

```python
for i in range(100):
    c0, c1 = (1,1)
    eps = np.sqrt(np.finfo(float).eps)
    grad = approx_fprime(xn, f, [eps, eps], c0, c1)
    xn = xn - learning_rate*grad
    yn = f(xn,c0, c1)
    Y[yn] = xn
Y
```

```python
print("минимум: ", min(Y.items())[1])
print("значение минимума: ", min(Y.items())[0])
```
